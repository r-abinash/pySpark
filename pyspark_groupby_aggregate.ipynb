```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('GroupBy_Aggregate').getOrCreate()
spark
```

```python
df = spark.read.csv('employee_data_two.csv',header=True,inferSchema=True)
df.show()
```

### GroupBy and Aggregation
 

```python
df.printSchema()
```

```python
# Apply groupBy by some column name.
df.groupBy('Name') # Its just return grouped data.
```

```python
# Applu aggregate function.
# grouped to find total salary based on name
df.groupBy('Name').sum() # sum will works on only with int columns not for string.ir return dataframe.
```

```python
df.groupBy('Name').sum().show()
```

```python
df.groupBy('Department').sum().show()
```

```python
# mean salary provided each department
df.groupBy('Department').mean().show()
```

```python
df.groupBy('Department').count().show()
```

```python
df.agg({'Salary':'Sum'}).show()
```
