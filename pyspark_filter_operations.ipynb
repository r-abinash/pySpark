```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Filter_operation').getOrCreate()
spark
```

```python
df = spark.read.csv('filter.csv',header=True,inferSchema=True)
df.show()
```

### method 1

```python
# filter based on some condition.
df.filter("Salary >= 20000").show() #return the entire dataframe.
```

```python
df.filter("Salary<20000").select(['Name','Age']).show() #do filter and return the selected columns only
# drawback is we can't apply multiple conditions by using logical operator.
```

### method 2

```python
# we can use logical operator.
df.filter(df['Salary']>=20000).show()
```

```python
df.filter((df['Salary']>=200) & (df['Age']>=23)).show() # while giving more conditions.ensure put eachone in seperate parantheses.
```

```python
df.filter((df['Salary']>=200) & 
          (df['Age']>=23)).select(['Name','Age']).show()
```

```python
df.filter(~(df['Salary']>=20000)).show() # not or inverse operation
```
